```python
import pandas as pd
import numpy as np
Encounters = pd.read_csv("Hospital_Project.csv")
Payer = pd.read_csv("Payers.csv")
df = Encounters.merge(Payer,left_on = "PAYER", right_on = "Payer_id", how = "left")

def calculate_risk_score(row):

    score = 0

    if row["ENCOUNTERCLASS"] == "urgentcare":
        score += 3
    elif row["ENCOUNTERCLASS"] == "emergency":
        score += 2
    elif row["ENCOUNTERCLASS"] == "ambulatory":
        score += 1
    elif row["ENCOUNTERCLASS"] == "outpatient":
        score += 1
    elif row["ENCOUNTERCLASS"] == "inpatient":
        score += 2

    if row["Encounter_Quality"] == "over 24 hours":
        score += 2

    if row["Payer_name"] in ["Anthem","Medicaid","NO INSURANCE"]:
        score += 1

    if row["Readmission_coulmn"] == "Yes":
        score += 3

    return score

df["Risk_Score"] = df.apply(calculate_risk_score, axis=1)

def risk_category(score):

    if score >= 7:
        return "High Risk"

    elif score >= 4:
        return "Medium Risk"

    else:
        return "Low Risk"
df["Risk_Category"] = df["Risk_Score"].apply(risk_category)
print(df[['ENCOUNTERCLASS', 
          'Encounter_Quality', 
          'Payer_name', 
          'Readmission_coulmn', 
          'Risk_Category', "Risk_Score"]].head(10))
print(df['Risk_Category'].value_counts())
print(df['Risk_Score'].value_counts())
df.to_csv('Healthcare_Risk_Assessment.csv', index=False)
print("Done!, file saved!")
```
